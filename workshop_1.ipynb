```python
from sqlalchemy import (create_engine, 
                        inspect, 
                        text, 
                        select, 
                        MetaData, 
                        Table, 
                        and_,
                        or_,
                        desc,
                        asc,
                        func,
                        )

engine_903 = create_engine("sqlite+pysqlite://///workspaces/Python_Learning/data/903_database.db")

connection = engine_903.connect()
inspection = inspect(connection)

inspection.get_table_names()
 
```

```python
engine_gravity = create_engine("sqlite+pysqlite://///workspaces/Python_Learning/data/gravity.db")

connection_gravity = engine_gravity.connect()
inspection_gravity = inspect(connection_gravity)

inspection_gravity.get_table_names()
 
```

```python
metadata_gravity = MetaData()

books = Table('book', metadata_gravity, autoload_with=engine_gravity)
print(books.name)
print(books.c.keys())
print(repr(books))
```

```python
with engine_gravity.connect() as con:
    stmt = "SELECT * FROM book"
    result_proxy = con.execute(text(stmt))
    results = result_proxy.fetchall()

first_row = results[0]
print(first_row)
print(first_row.title)
```

```python
with gravity_engine.connect() as con:
    stmt = "SELECT * FROM book"
    result_proxy = con.execute(text(stmt))
    results = result_proxy.fetchmany(size=10)

print(results)
```

```python
with engine_gravity.connect() as con:

    stmt = select(text('SELECT * FROM book'))
    # print(stmt)
    result_proxy = con.execute(stmt)
    result = result_proxy.fetchmany(size=5)

print(result)
```

```python
with engine_gravity.connect() as con:
    stmt = select(books)
    result = con.execute(stmt).fetchmany(size=5)

print(result)
```

```python
header = Table('header', metadata_gravity, autoload_with=engine_gravity)

with engine_gravity.connect() as con:
    stmt = select(header).where(or_(header.columns.publication_date == "1999",
                                    header.columns.publication_date == "2000"))
    # print(stmt)
    result_proxy = con.execute(stmt)
    results = result_proxy.fetchmany(size=10)

print(results)
```

```python
# select all of the book table where the publication_date is either 1999 or 2000

with engine_gravity.connect() as con:
    stmt = select(books).where(or_(books.columns.book_id == 1,
                                   books.columns.book_id == 2))
    
    result_proxy = con.execute(stmt)
    results = result_proxy.fetchmany(size=10)

print(results)
```
